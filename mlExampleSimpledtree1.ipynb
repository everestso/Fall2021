```
# Common imports
import numpy as np

# to make this notebook's output stable across runs
np.random.seed(42)

import matplotlib.pyplot as plt

import matplotlib.image as mpimg
import pydotplus

import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import math
```

```
import os
from google.colab import drive

drive.mount('/content/gdrive')
```

```
datasetsDir = "/content/gdrive/My Drive/csuf/Resources/Datasets/"
csv_path = os.path.join(datasetsDir, "MlSimpleTest1 - Sheet4.csv")
data = pd.read_csv(csv_path)
print(data)
```

```
X = data.drop("y", axis=1)
Y=data['y'].values

classifier1 = DecisionTreeClassifier(criterion="gini")
classifier2 = DecisionTreeClassifier(criterion="entropy")
classifier1.fit(X,Y)
classifier2.fit(X,Y)
```

```
dotfile = open("dtree2.dot", 'w')
export_graphviz(classifier1, out_file = dotfile, feature_names = X.columns)
dotfile.close()

graph = pydotplus.graph_from_dot_file("dtree2.dot")
graph.write_png('test.png')
img=mpimg.imread('test.png')
fig = plt.figure(figsize=(7, 7)) 
plt.axis("off")
plt.imshow(img)
plt.show()
```

```
dotfile = open("dtree2.dot", 'w')
export_graphviz(classifier2, out_file = dotfile, feature_names = X.columns)
dotfile.close()

graph = pydotplus.graph_from_dot_file("dtree2.dot")
graph.write_png('test.png')
img=mpimg.imread('test.png')
fig = plt.figure(figsize=(7, 7)) 
plt.axis("off")
plt.imshow(img)
plt.show()
```
