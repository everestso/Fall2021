```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn import datasets

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


iris = datasets.load_iris()
X = iris.data[:, :4]  # we only take the first two features.
R = iris.target

plt.plot(X[:50,0], X[:50,2], 'r+')
plt.plot(X[51:100,0], X[51:100,2], 'go')
plt.plot(X[101:150,0], X[101:150,2], 'b*')

plt.show()
plt.scatter(X[:,2], X[:,3], c=R)
plt.show()
```

```python
Xpca = PCA(n_components=2)
Xred = pca.fit_transform(X)
print(pca.explained_variance_ratio_, sum(pca.explained_variance_ratio_))
print (pca.components_[0,:])
print (pca.components_[1,:])
print (X[0,:],Xred[0,:])
```

```python
F= lambda w, a: w[0]*a[0] + w[1]*a[1] + w[2]*a[2] + w[3]*a[3]
Xcen = X - X.mean(axis=0)
print (X.mean(axis=0))
print (X[0,:])
print ("X=", Xcen[0,:])
print (pca.components_[0,:])
print (F(pca.components_[0,:],Xcen[0,:]))
print (F(pca.components_[1,:],Xcen[0,:]))
print (Xred[0,:])
```

```python
plt.plot(Xred[:50,0], Xred[:50,1], 'r+')
plt.plot(Xred[51:100,0], Xred[51:100,1], 'go')
plt.plot(Xred[101:150,0], Xred[101:150,1], 'b*')

plt.show()
```

```python
Xtrain = list(zip(X[: , 0], X[:,1]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))

Xtrain = list(zip(X[: , 0], X[:,2]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))

Xtrain = list(zip(X[: , 0], X[:,3]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))

Xtrain = list(zip(X[: , 1], X[:,2]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))

Xtrain = list(zip(X[: , 1], X[:,3]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))

Xtrain = list(zip(X[: , 2], X[:,3]))
print (Xtrain)
lr = LogisticRegression(solver='lbfgs')
lr.fit(Xtrain, R)
predict = lr.predict(Xtrain)
print(accuracy_score(predict, R))
```
