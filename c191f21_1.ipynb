```python
### Includes
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
Weights = [5, 2] # F(X) = 5x_0 + 2x_1
F = lambda W, X: sum([w*x for w,x in zip(W,X)])

dataX1 = range(-10, 11, 1)
dataX0 = [1]*len(dataX1)

data = (list(zip(dataX0, dataX1)))
print (data)
Y = [F(Weights, X) for X in data]
print (Y)
```

```python
plt.plot(dataX1, Y)
plt.show()
```

```python
dataX2 = [x**2 for x in dataX1]
data = list(zip(dataX0, dataX1, dataX2))
print (data)

Weights = [0, 0, 3]
Y = [F(Weights, X) for X in data]
plt.plot(dataX1, Y, "g")
plt.show()
```

# Now Learning w/ Line

```python
Weights = [5, 2] # F(X) = 5x_0 + 2x_1
F = lambda W, X: sum([w*x for w,x in zip(W,X)])

dataX1 = range(-10, 11, 1)
dataX0 = [1]*len(dataX1)

data = (list(zip(dataX0, dataX1)))
print (data)
Y = [F(Weights, X) for X in data]
print (Y)
```

```python
## Add Noise
import random
import numpy as np
EPS = 100
yNoise = [y+random.random()*EPS-EPS/2 for y in Y]

plt.plot(dataX1, Y, "g")
plt.plot(dataX1, yNoise, "g+")
plt.show()
```

```python
cv =np.cov(dataX1, yNoise, bias=True, rowvar=False)
w1 = cv[1][0]/cv[0][0]
print(w1)
```

```python
Xbar, Ybar = np.mean(dataX1), np.mean(yNoise)
w0 = Ybar - w1*Xbar
print(w0,w1)
```

```python
yPredicted = [F([w0,w1], x) for x in data]
plt.plot(dataX1, Y, "b")
plt.plot(dataX1, yPredicted, "g")
plt.plot(dataX1, yNoise, "g+")
plt.show()
```

```python
data = [((1, -5), -5), ((1, -4), -3), ((1, -3), -1), ((1, -2), 1), ((1, -1), 3), ((1, 0), 5), ((1, 1), 7), ((1, 2), 9), ((1, 3), 11), ((1, 4), 13), ((1, 5), 15)]
print(data)
X = [d for d,_ in data]
Y = [y for _, y in data]
print(X)
print(Y)
```

```python
alpha=0.01
model=[0, 0]
```

```python
# Calculate Predicted

yPredicted = [F(model, x) for x in X]
print(yPredicted)
```

```python
# Evaluate Model
Error = sum([(r-y)**2.0 for r,y in zip(Y, yPredicted)])
print(Error)
```

```python
deltaX0 = 1/len(Y)*sum( [(-2*(r-y)*x[0]) for r,y,x in zip(Y, yPredicted, X)] )
deltaX1 = 1/len(Y)*sum( [(-2*(r-y)*x[1]) for r,y,x in zip(Y, yPredicted, X)] )

print(deltaX0)
print(deltaX1)
```

```python
model = [model[0]-alpha*deltaX0, model[1]-alpha*deltaX1]
print(model)
```
