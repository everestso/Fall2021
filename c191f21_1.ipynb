# Now Learning w/ Line

```python
Weights = [5, 2] # F(X) = 5x_0 + 2x_1
F = lambda W, X: sum([w*x for w,x in zip(W,X)])

dataX1 = range(-10, 11, 1)
dataX0 = [1]*len(dataX1)

data = (list(zip(dataX0, dataX1)))
print (data)
Y = [F(Weights, X) for X in data]
print (Y)
```

```python
## Add Noise
import random
import numpy as np
EPS = 100
yNoise = [y+random.random()*EPS-EPS/2 for y in Y]

plt.plot(dataX1, Y, "g")
plt.plot(dataX1, yNoise, "g+")
plt.show()
```

```python
cv =np.cov(dataX1, yNoise, bias=True, rowvar=False)
w1 = cv[1][0]/cv[0][0]
print(w1)
```

```python
Xbar, Ybar = np.mean(dataX1), np.mean(yNoise)
w0 = Ybar - w1*Xbar
print(w0,w1)
```

```python
yPredicted = [F([w0,w1], x) for x in data]
plt.plot(dataX1, Y, "b")
plt.plot(dataX1, yPredicted, "g")
plt.plot(dataX1, yNoise, "g+")
plt.show()
```

### Cubic function

```python
Weights = [0.5, 0.2, 0.25, 0.5] # F(X) = 0.5x_0 + 0.2x_1, 0.25x_1^2 + 0.5x_1^3
F = lambda W, X: sum([w*x for w,x in zip(W,X)])

dataX1 = [i*0.01 for i in range(-500, 501, 10)]
dataX0 = [1]*len(dataX1)
dataX2 = [x**2 for x in dataX1]
dataX3 = [x**3 for x in dataX1]

data = list(zip(dataX0, dataX1, dataX2, dataX3))

Y = [F(Weights, X) for X in data]
TrainingData1 = list(zip(data, Y)) 
```

```python
plt.plot(dataX1, Y)
plt.show()
```

```python
data = TrainingData1
X = [(d[0], d[1], d[2], d[3]) for d,_ in data]
Y = [y for _, y in data]

alpha=0.0001
model=[0, 0, 0, 0]
Errors = []
iteration = 0
```

```python
yPredicted = [F(model, x) for x in X]
Error = sum([(r-y)**2.0 for r,y in zip(Y, yPredicted)])
if iteration%1==0:
  print("Iteration: ", iteration, "Error=", Error)

deltaX0 = 1/len(Y)*sum( [(-2*(r-y)*x[0]) for r,y,x in zip(Y, yPredicted, X)] )
deltaX1 = 1/len(Y)*sum( [(-2*(r-y)*x[1]) for r,y,x in zip(Y, yPredicted, X)] )
deltaX2 = 1/len(Y)*sum( [(-2*(r-y)*x[2]) for r,y,x in zip(Y, yPredicted, X)] )
deltaX3 = 1/len(Y)*sum( [(-2*(r-y)*x[3]) for r,y,x in zip(Y, yPredicted, X)] )

print(deltaX0, deltaX1, deltaX2, deltaX3)

model = [model[0]-alpha*deltaX0, model[1]-alpha*deltaX1, model[1]-alpha*deltaX2, model[1]-alpha*deltaX3]
print(model)
iteration += 1
```

```python
for _ in range(100):
  yPredicted = [F(model, x) for x in X]
  Error = sum([(r-y)**2.0 for r,y in zip(Y, yPredicted)])
  if iteration%10==0:
    print("Iteration: ", iteration, "Error=", Error)

  deltaX0 = 1/len(Y)*sum( [(-2*(r-y)*x[0]) for r,y,x in zip(Y, yPredicted, X)] )
  deltaX1 = 1/len(Y)*sum( [(-2*(r-y)*x[1]) for r,y,x in zip(Y, yPredicted, X)] )
  deltaX2 = 1/len(Y)*sum( [(-2*(r-y)*x[2]) for r,y,x in zip(Y, yPredicted, X)] )
  deltaX3 = 1/len(Y)*sum( [(-2*(r-y)*x[3]) for r,y,x in zip(Y, yPredicted, X)] )

  model = [model[0]-alpha*deltaX0, model[1]-alpha*deltaX1, model[1]-alpha*deltaX2, model[1]-alpha*deltaX3]
  iteration += 1
```
